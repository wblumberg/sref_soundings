```python
from netCDF4 import Dataset
import netCDF4
import numpy as np
from pylab import *
from datetime import datetime

run_hour = '09'
yyyymmdd = '20131205'
sref_data = 'http://nomads.ncep.noaa.gov:9090/dods/sref/sref' + yyyymmdd + '/'
latitude = 35.2167
longitude = -97.4167
members = 'sref_memberlist'
time_idx = 6

member_file = open(members, 'r')
#print member_file.readlines()

#hghprs
#
```

```python
firstrun = 0
lat_idx = 0
lon_idx = 0
pres = None
temperatures = []
heights = []
colors = []
labels= []
rhs = []
time = None
```

```python
for line in member_file.readlines():
    membername = line.rstrip()
    membername, color, label = membername.split(',')
    dap_link = sref_data + membername + run_hour + 'z'
    print dap_link
    d = Dataset(dap_link)
    colors.append(color)
    labels.append(label)

    #print d.variables.keys()
    if firstrun == 0:
        times = netCDF4.num2date(d.variables['time'],d.variables['time'].units)
        time = times[time_idx]
        lat = d.variables['lat'][:]
        lon = d.variables['lon'][:]
        pres = d.variables['lev'][:]
        lat_idx = np.argmin(np.abs(latitude - lat))
        lon_idx = np.argmin(np.abs(longitude - lon))
    #height = d.variables['hgtprs'][:, :, lat_idx, lon_idx]
    #height = height * 9.81
    temp = d.variables['tmpprs'][:, :, lat_idx, lon_idx]
    rhs.append(d.variables['rhprs'][:, :, lat_idx, lon_idx])
    print d.variables['pressfc'][:,lat_idx, lon_idx]
    temperatures.append(temp)
    #heights.append(height)
    d.close()
```

```python
temperatures = np.asarray(temperatures)
#heights = np.asarray(heights)
temperatures.shape
labels = np.asarray(labels)
rhs = np.asarray(rhs)
print labels
```

```python
def plotProfile(temps, pres, rh, time, col, lab):
    figure(figsize=(12,7))
    subplot(121)
    rc('xtick', labelsize=18) 
    rc('ytick', labelsize=18)
    for t,c,l in zip(temps, col, lab):
        #print len(np.where(t > 273.15)[0])
        #print t,l
        if len(np.where(t > 273.15)[0]) == 0:
            plot(t-273.15, pres, c='#' + c, lw=3, label=l)
        else:
            plot(t-273.15, pres, c='#' + c, lw=2, alpha=.5,label=l)
    ylim(400,980)
    xlim(-20,5)
    suptitle(datetime.strftime(times[0], '%Y-%m-%d %HZ') + " SREF OUN Profiles \nValid " + datetime.strftime(time, '%Y-%m-%d %HZ'), fontsize=16) 
    ylabel("Pressure [mb]", fontsize=18)
    xlabel("Temperature [C]", fontsize=18)
    grid()
    axvline(x=0, c='k', lw=4)
    gca().invert_yaxis()
    
    subplot(122)
    for r,c,l in zip(rh, col, lab):
        #print len(np.where(t > 273.15)[0])
        #print t,l
        plot(r, pres, c='#' + c, lw=3, label=l)
    ylim(400,980)
    #xlim(-12,5)
    #title("2013-12-04 21Z SREF OUN Temp Profiles \nValid 12-06 9Z", fontsize=16) 
    #ylabel("Pressure [mb]", fontsize=18)
    xlabel("Relative Humidity [%]", fontsize=18)
    grid()
    gca().invert_yaxis()
    
    legend(loc='center left', bbox_to_anchor=(1,.5),ncol=1, fontsize=11.5)
    #show()
    savefig(datetime.strftime(time, '%Y%m%d.%HZ.OUN.png'))
    clf()
    #import sys
    #sys.exit(1)
```

```python
temperatures = np.reshape(temperatures, (temperatures.shape[1], temperatures.shape[0], temperatures.shape[2]))
rhs = np.reshape(rhs, (rhs.shape[1], rhs.shape[0], temperatures.shape[2]))
```

```python
for t, r, ti in zip(temperatures, rhs, times):
    plotProfile(t, pres, r, ti, colors, labels)
    #show()
```
